# 1. Review: The building blocks of Python code

The two building blocks of Python code are *expressions* and *statements*.  An **expression** is a piece of code that

* is self-contained, meaning it would make sense to write it on a line by itself, and
* usually has a value.

    3

| Function | Description                                                   |

## 2.1. String Methods

Strings can be transformed using **methods**, which are functions that work on objects such as strings or tables. One example is the `replace` method, which replaces all instances of some part of a string with some alternative. 

A method is invoked on a string by placing a `.` after the string value, then the name of the method, and finally parentheses containing its arguments. Here's a sketch, where the `<` and `>` symbols aren't part of the syntax; they just mark the boundaries of sub-expressions.

    <expression that evaluates to a string>.<method name>(<argument>, <argument>, ...)

Try to predict the output of these examples, then execute them.

```python
# Replace one letter
'Hello'.replace('o', 'a')
```

```python
# Replace a sequence of letters, which appears twice
'hitchhiker'.replace('hi', 'ma')
```

Once a name is bound to a string value, methods can be invoked on that name as well. The name's value doesn't change in this case, so a new name is needed to capture the result. 

In the example below, doing "sharp.replace(...)" doesn't change the value of sharp, so a new name "hot" saves the new result. 

```python
sharp = 'edged'
hot = sharp.replace('ed', 'ma')
print('sharp:', sharp)
print('hot:', hot)
```

As we have seen, you can call functions on the results of other functions.  For example,

    max(abs(-5), abs(3))

has value 5.  Similarly, you can invoke methods on the results of other method (or function) calls.

```python
# Calling replace on the output of another call to
# replace
'train'.replace('t', 'ing').replace('in', 'de')
```

## 2.4 Exercising control over our code

Python is capable of logic and making decisions, but only in a very binary way. We can do this by using booleans, which are values designated by `True` or `False` (note the capitalization). These are very specific values: `True` evaluates to 1, while `False` evaluates to 0. We can create these by simply writing out the word, but a more useful approach is to make comparisons. Python has a variety of operators that allow us to compare expressions, such as == ("equal to"), > ("greater than"), and <= ('less than or equal to"). If the expression comparison is True, then Python will return True.

We can also make compound statements using `and` and `or`. In the case of `and`, both the statement on the left `and` the statement on the right must be True for Python to return True. In the case of `or`, either the statement on the left `or` the statement on the right must be True for Python to return True. For example:

```python
## What would this cell return? Think about it before running the cell.
a, b = 3, 5
x, y = 9, 4
x > y or a < b
```

Booleans are helpful for when we want to make decisions in our code, such as when a specific condition is True. We can do this by using if/else statements, which have the following structure as seen in the coding cell below.

Notice how the if/else statement begins with an `if`, followed by some amount of `elif`, and ending with an `else`, with all code after each if/else statement indented. This means that, when the expression is True, the indented code under that line will be run, and the rest will be ignored. The `if` designates the first condition to begin. Each `elif` designates other possible conditions, and `else` will be what the if/else statement returns if every other condition is False. You can have as many elifs as you would like, and you do not necessarily need to have an else statement at the end.   

```python
if a < 0:
    print("a is negative!")
elif b < 0: 
    print("b is negative!")
elif x > 0:
    print("x is positive!")
elif y > 0:
    print("y is positive!")
else:
    print("neither a and b are negative, nor are x and y positive")
```

Rather than copy-pasting, Python allows us to **import** other code, creating a **module** that contains all of the names created by that code. You can think of a **module** as a storage space for the pieces of code you might want to use; having it in your code ensures that Python has access to the code you want to use.

Python includes many useful modules that are just an `import` away.  An example is the `math` module, which is extremely useful in computing mathematical expressions in Python.
