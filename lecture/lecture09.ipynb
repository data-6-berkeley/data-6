## `group`

```python
## read_table primarily uses CSVs, but also takes direct links to a file
marvel = Table.read_table("https://github.com/fivethirtyeight/data/raw/master/comic-characters/marvel-wikia-data.csv")
marvel
```

```python
## Find count of alignment of characters, vs frequency
total_chars = sum(marvel.group('ALIGN').column(1))
total_chars
grouped_marvel = marvel.group("ALIGN")

grouped_marvel.with_columns("Proportion", grouped_marvel.column("count") / total_chars)
```

## `barh`

```python
## specifying a quantitative variable
minard_with_percents.barh("City", "Percent surviving")
```

```python
# creating a chart of counts
marvel.group("HAIR").barh("HAIR", "count")
```

```python
# creating a chart of frequencies
by_hair = marvel.group("HAIR")
hair_percents = 100*by_hair.column("count") / sum(by_hair.column("count"))
with_hair_percents = by_hair.with_columns("percent", hair_percents)
with_hair_percents
```

```python
with_hair_percents.barh("HAIR", "percent")
```
